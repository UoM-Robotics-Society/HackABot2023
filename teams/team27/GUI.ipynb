```python

# arduino_LED_user.py

import serial
from PyQt5.QtWidgets import *
from PyQt5.QtGui import QKeySequence
from PyQt5 import QtCore
import time
import sys

MIN_THETA = -255
MAX_THETA = 255
MIN_VELOCITY = -255
MAX_VELOCITY = 255
INCREMENT = 40
ser = serial.Serial('COM11', 9600)

class Window(QWidget):
    def __init__(self):
        QWidget.__init__(self)
        layout = QHBoxLayout()
        self.setLayout(layout)
        self.dial = QDial()
        self.dial.setMinimum(MIN_THETA)
        self.dial.setMaximum(MAX_THETA)
        self.dial.setValue(0)
        self.dial.valueChanged.connect(self.send_to_ser)
        layout.addWidget(self.dial)
        QShortcut(QKeySequence("D"), self).activated.connect(lambda: self.on_key_theta(-INCREMENT))
        QShortcut(QKeySequence("A"), self).activated.connect(lambda: self.on_key_theta(INCREMENT))
        self.sp = QSlider(orientation=QtCore.Qt.Vertical)
        self.sp.setMinimum(MIN_VELOCITY)
        self.sp.setMaximum(MAX_VELOCITY)
        self.sp.valueChanged.connect(self.send_to_ser)
        layout.addWidget(self.sp)
        QShortcut(QKeySequence("W"), self).activated.connect(lambda: self.on_key_velocity(INCREMENT))
        QShortcut(QKeySequence("S"), self).activated.connect(lambda: self.on_key_velocity(-INCREMENT))
        
        self.dial2 = QDial()
        self.dial2.setMinimum(MIN_THETA)
        self.dial2.setMaximum(MAX_THETA)
        self.dial2.setValue(0)
        self.dial2.valueChanged.connect(self.goalie_send_to_ser)
        layout.addWidget(self.dial2)
        QShortcut(QKeySequence("L"), self).activated.connect(lambda: self.on_key_theta2(-INCREMENT))
        QShortcut(QKeySequence("J"), self).activated.connect(lambda: self.on_key_theta2(INCREMENT))
        self.sp2 = QSlider(orientation=QtCore.Qt.Vertical)
        self.sp2.setMinimum(MIN_VELOCITY)
        self.sp2.setMaximum(MAX_VELOCITY)
        self.sp2.valueChanged.connect(self.goalie_send_to_ser)
        layout.addWidget(self.sp2)
        QShortcut(QKeySequence("I"), self).activated.connect(lambda: self.on_key_velocity2(INCREMENT))
        QShortcut(QKeySequence("K"), self).activated.connect(lambda: self.on_key_velocity2(-INCREMENT))
    

    def on_key_theta(self, val_change):
        new_val = self.dial.value()+val_change
        new_val = min(new_val, MAX_THETA)
        new_val = max(MIN_THETA, new_val)
        self.dial.setValue(new_val)
    
    def on_key_velocity(self, val_change):
        new_val = self.sp.value()+val_change
        new_val = min(new_val, MAX_VELOCITY)
        new_val = max(MIN_VELOCITY, new_val)
        self.sp.setValue(new_val)
        
    def on_key_theta2(self, val_change):
        new_val = self.dial2.value()+val_change
        new_val = min(new_val, MAX_THETA)
        new_val = max(MIN_THETA, new_val)
        self.dial2.setValue(new_val)
    
    def on_key_velocity2(self, val_change):
        new_val = self.sp2.value()+val_change
        new_val = min(new_val, MAX_VELOCITY)
        new_val = max(MIN_VELOCITY, new_val)
        self.sp2.setValue(new_val)
    
    def send_to_ser(self):
        to_write = f"a{self.dial.value()},{self.sp.value()}"
        ser.write(to_write.encode())
        print(to_write)
    
    def goalie_send_to_ser(self):
        to_write = f"g{self.dial2.value()},{self.sp2.value()}"
        ser.write(to_write.encode())
        print(to_write)
        
app = QApplication(sys.argv)
app.setStyle('QtCurve')
screen = Window()
screen.show()
sys.exit(app.exec_())
```
